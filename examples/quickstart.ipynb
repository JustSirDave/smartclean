```python
import pandas as pd
import smartclean as sc
```

```python
# Load data
df = sc.read(r"C:\Users\David\Desktop\Models\smartclean-root\docs\titanic.csv")
print("Original shape:", df.shape)
print("\nFirst look:")
print(df.head())
print("\nMissing values:")
print(df.isna().sum())
```

```python
df.head(10)
```

```python
# Profile the dataset
print("\n")
result = sc.profile(df)
print(result.summary())
```

```python
# Auto clean
print("\n" + "="*60)
print("AUTO CLEAN")
print("="*60)
clean_df, report = sc.auto_clean(df, return_report=True)
report.summary()
```

```python
# ── 4. Inspect the result ────────────────────────────────────────────────────
print("Clean shape:", clean_df.shape)
print("\nClean columns:", list(clean_df.columns))
print("\nMissing values after cleaning:\n")
print(clean_df.isna().sum())
```

```python
clean_df.head(10)
```

```python
# 5. Manual API
print("\n" + "="*60)
print("MANUAL CHAINED API")
print("="*60)
cleaner = sc.Cleaner(df)
manual_df = (
    cleaner
    .clean_columns()
    .fix_types()
    .handle_missing()
    .remove_duplicates()
    .clean_text()
    .remove_outliers()
    .output()
)
cleaner.get_report().summary()
print("Manual clean shape:", manual_df.shape)
```
